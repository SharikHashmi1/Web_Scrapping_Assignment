# 2. In the above question, now scrape the following details of each product listed in first 3 pages of your 
search results and save it in a data frame and csv. In case if any product has less than 3 pages in search 
results then scrape all the products available under that product name. Details to be scraped are: "Brand 
Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and 
“Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“.

```python
# get product url
product_urls = []
for i in range(0,3): 
    product_url = driver.find_elements_by_xpath("//a[@class='a-link-normal a-text-normal']")
    for i in product_url:
        product_urls.append(i.get_attribute('href'))
    try:
        next_page=driver.find_element_by_xpath("//a[@class='s-pagination-item s-pagination-next s-pagination-button s-pagination-separator']").get_attribute('href')
        next_page.click(_)
        driver.refresh()   
    except:
        break
```

```python
# Creating blank lists to store data after fetching
product_names = []
brand_names = []
ratings = []
no_ratings = []
prices = []
exchanges = []
del_dates = []
avails = []
other_details = []
links = []
```

```python
for link in product_urls:
    driver.get(link)
           
    try:
        name = driver.find_element_by_xpath("//span[@id='productTitle']")
        product_names.append(name.text)
    except NoSuchElementException:
        product_names.append('-')
        
    try:
        brand = driver.find_element_by_xpath("//a[@id='bylineInfo']")
        brand_names.append(brand.text)
    except NoSuchElementException:
        brand_names.append('-')
        
    try:
        rats = driver.find_element_by_xpath("//span[@class='a-size-base a-nowrap']")
        ratings.append(rats.text)
    except NoSuchElementException:
        ratings.append('-')
        
    try:
        n_r = driver.find_element_by_xpath("//span[@id='acrCustomerReviewText']")
        no_ratings.append(n_r.text)
    except NoSuchElementException:
        no_ratings.append('-')
        
    try:
        price = driver.find_element_by_xpath("//span[@id='priceblock_ourprice']")
        prices.append(price.text)
    except NoSuchElementException:
        prices.append('-')
        
    try:
        excs = driver.find_elements_by_xpath("//div[@class='a-section a-spacing-none icon-content']")
        try: 
            exchanges.append(excs[1].text)
        except:
            continue
    except NoSuchElementException:
        exchanges.append('-')
        
    try:
        dels = driver.find_element_by_xpath("//div[@id='ddmDeliveryMessage']/b")
        del_dates.append(dels.text)
    except NoSuchElementException:
        del_dates.append('-')
        
    try:
        ava = driver.find_element_by_xpath("//span[@class='a-size-medium a-color-price'] | //span[@class='a-size-medium a-color-success'] ")
        avails.append(ava.text)
    except NoSuchElementException:
        avails.append('-')
        
    try:
        det = driver.find_element_by_xpath("//div[@id='productDescription']")
        other_details.append(det.text)
    except NoSuchElementException:
        other_details.append('-')
        
    try:
        links.append(link)
    except:
        links.append('-')
```

```python
# Creating a DATA FRAME to store fetched data
Guitar = pd.DataFrame({})
Guitar['Name'] = product_names
Guitar['Brand'] = brand_names
Guitar['Rating'] = ratings 
Guitar['No. ratings'] = no_ratings 
Guitar['Price'] = prices
Guitar['Exchange'] = exchanges
Guitar['Delivery Date'] = del_dates 
Guitar['Availability'] = avails 
Guitar['Details'] = other_details
Guitar['Product_Url'] = links
Guitar
```

# 3. Write a python program to access the search bar and search button on images.google.com and scrape 100 images each for keywords ‘fruits’, ‘cars’ and ‘Machine Learning’.

```python
import pandas as pd
import selenium
from selenium import webdriver
import time
from selenium.common.exceptions import NoSuchElementException, StaleElementReferenceException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.keys import Keys
```

```python
# Connecting to the webdriver
driver=webdriver.Chrome(r'C:/Users/hp/Downloads/chromedriver_win32/chromedriver.exe')
# Getting the webpage of mentioned url 
url = "https://images.google.com/"
driver.get(url)
```

```python
# Creating a function which takes the necessary input as argument
def get_urls(keywords):
        driver.get(url)

        # Locating the search bar and click search button
        search_bar=driver.find_element_by_xpath("//input[@class='gLFyf gsfi']")
        search_bar.send_keys(keywords)
        button=driver.find_element_by_xpath("//button[@class='Tg7LZd']")
        button.click()
        time.sleep(3)
        
        # loading few pages to scrape images
        for k in range(10):
            driver.find_element_by_xpath("//a[@class='wXeWr islib nfEiy']").send_keys(Keys.PAGE_DOWN)
        time.sleep(5)
        picture = [image for image in driver.find_elements_by_xpath("//a[@class='wXeWr islib nfEiy']")[:100]]
        image_urls=[]
        print(len(picture))
        if(len(picture)==100):
            for i in picture:
                try:
                    i.click()
                    image_urls.append(i.get_attribute('href'))
                except:
                    image_urls.append('Not Available')
            return image_urls
        else:
            print('Image Not Available')
```

```python
#Let's scrape the url's of all images
fruits=get_urls('Fruits')
cars=get_urls('Cars')
machine_learning=get_urls('Machine Learning')
```

```python
# Let's create a DataFrame for our data
Image_URL = pd.DataFrame({})
Image_URL['Fruits'] = fruits
Image_URL['Cars'] = cars
Image_URL['Machine Learning'] = machine_learning
Image_URL
```

```python
import pandas as pd
import selenium
from selenium import webdriver
import time
from selenium.common.exceptions import NoSuchElementException, StaleElementReferenceException
from selenium.webdriver.support.ui import WebDriverWait
```

```python
# Connecting to the webdriver
driver = webdriver.Chrome(r'C:/Users/hp/Downloads/chromedriver_win32/chromedriver.exe')
# Getting the webpage of mentioned url 
url = "https://www.youtube.com/"
driver.get(url)
```

```python
#lets click on a video
clk_video = driver.find_element_by_xpath("//a[@id='video-title-link']").click()
time.sleep(8)
```

```python
#lets find required data
hostel_name = []
distance = []
pvt_prices = []
dorms_price = []
rating = []
reviews = []
over_all = []
facilities = []
description =[]
Product_url = []
```

```python
for i in driver.find_elements_by_xpath("//div[@class = 'pagination-item pagination-next']"):
    i.click()
    time.sleep(2)
    
    #Scraping hostel name
    try:
        name = driver.find_elements_by_xpath("//h2[@class='Astor Museum Inn ']")
        for i in name:
            hostel_name.append(i.text)
    except NoSuchElementException:
        hostel_name.append('-')
        
    # scraping the distance from city centre    
    try:
        dist = driver.find_elements_by_xpath("//span[@class='description']")
        for i in dist:
            Distance.append(i.text.replace('Hostel - ',''))
    except NoSuchElementException:
        Distance.append('-')
        
    # scraping the privates from price
    for i in driver.find_elements_by_xpath("//div[@class='prices-col']"):
        try:
            pvt_price = driver.find_element_by_xpath("//a[@class='prices']//div[1]//div")
            Pvt_prices.append(pvt_price.text)
        except NoSuchElementException:
            Pvt_prices.append('-')
            
    # scraping the dorms from price
    for i in driver.find_elements_by_xpath("//div[@class='prices-col']"):
        try:
            dorms = driver.find_element_by_xpath("//a[@class='prices']//div[2]//div")
            Dorms_price.append(dorms.text)
        except NoSuchElementException:
            Dorms_price.append('-')
            
        time.sleep(2)        
    
    # scraping the facilities
    try:
        faci = driver.find_elements_by_xpath('//div[@class="facilities-label facilities"]')
        for i in faci:
            Facilities.append(i.text)
    except NoSuchElementException:
        Facilities.append('-')
        
    time.sleep(2)
    
    # scraping the urls of each hostel
    pro_url = driver.find_elements_by_xpath("//div[@class='prices-col']//a[2]")
    for i in pro_url:
        Product_url.append(i.get_attribute('href'))
    
    time.sleep(2)

for i in Product_url:
    driver.get(i)
    time.sleep(2)
    
    # clicking on the show more button for description
    try:
        driver.find_element_by_xpath("//a[@class='toggle-content']").click()
        time.sleep(2)
    except NoSuchElementException:
        pass
    
    # scraping the ratings for the hostels
    try:
        rating = driver.find_element_by_xpath("//div[@class='score orange big']")
        Rating.append(rating.text)
    except NoSuchElementException:
        Rating.append('-')
    
    time.sleep(2)
    
    # scraping the total reviews        
    try:
        rev = driver.find_element_by_xpath("//div[@class='reviews']")
        Reviews.append(rev.text.replace('Total Reviews',''))
    except NoSuchElementException:
        Reviews.append('-')
    
    time.sleep(2)
    
    # scraping the overall review
    try:
        over_rev = driver.find_element_by_xpath("//div[@class='keyword']//span")
        Over_all.append(over_rev.text)
    except NoSuchElementException:
        Over_all.append('-')
        
        time.sleep(2)
    
    # scraping the property description 
    try:
        des = driver.find_element_by_xpath("//div[@class='content']")
        Description.append(des.text)
    except NoSuchElementException:
        Over_all.append('-')
```
